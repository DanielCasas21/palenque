```python
import pandas as pd
import csv

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv(csv_file)

# Randomly sample 100 rows without replacement
sample_df = df.sample(n=100)

# Extract the information in column 'c'
column_filtered = sample_df['filtered']

# Save the extracted data to a text file without additional quotes
column_filtered.to_csv('sampled_column_filtered.txt', index=False, header=False, quoting=csv.QUOTE_NONE, escapechar='\\')

# Open the text file and replace '\,' with ','
with open('sampled_column_filtered.txt', 'r') as file:
    content = file.read()

# Replace the escape sequence '\,' with ','
content = content.replace('\\,', ',')

# Save the modified content back to the file
with open('sampled_column_filtered.txt', 'w') as file:
    file.write(content)
```
