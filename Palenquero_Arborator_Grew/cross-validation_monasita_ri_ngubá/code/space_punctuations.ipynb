```python
def add_space_before_punctuation(file_path):
    try:
        with open(file_path, 'r') as file:
            text = file.read()

        modified_text = ''
        for char in text:
            if char in ['.', ',', '!', '?', ':', ';']:
                modified_text += ' ' + char
            else:
                modified_text += char

        with open(file_path, 'w') as file:
            file.write(modified_text)

        print("Spaces added before punctuation marks successfully.")
    except FileNotFoundError:
        print("File not found.")

# Example usage:
file_path = txt_file  # Replace 'your_file.txt' with the path to your text file
add_space_before_punctuation(file_path)
```
