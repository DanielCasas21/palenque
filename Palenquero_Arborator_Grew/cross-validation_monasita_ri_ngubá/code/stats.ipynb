```python
# @title UPOSTAG Heat Maps

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plot_filtered_dfs = [filtered_df1, filtered_df2, filtered_df3, filtered_df4]
names_filtered_dfs = ["Palenquero_Non_Sequential", "Palenquero_Words_7", "Palenquero_10", "Palenquero_Jundo"]

for i, (f, name) in enumerate(zip(plot_filtered_dfs, names_filtered_dfs), start=1):
    # Create a new figure and axis object
    fig, ax = plt.subplots(figsize=(8, 8))

    # Set the title for the current plot
    ax.set_title(f"Filtered DataFrame: {name}", pad=15)

    # Create a copy of the DataFrame to avoid SettingWithCopyWarning
    f = f.copy()

    # Replace values in 'root_head_match' column
    f['root_head_match'] = f['root_head_match'].replace({0: 'no-match', 1: 'match'})

    # Create the 2D histogram DataFrame
    df_2dhist = pd.DataFrame({
        x_label: grp['UPOSTAG'].value_counts()
        for x_label, grp in f.groupby('root_head_match')
    }).fillna(0)  # Fill NaN with 0 for empty cells

    # Plot the heatmap
    sns.heatmap(df_2dhist, cmap='viridis', annot=True, fmt='g', ax=ax, linewidths=1, linecolor='white')  # Add lines separating sections

    # Set the labels
    ax.set_xlabel('Root-Head Match', labelpad=15)
    ax.set_ylabel('UPOSTAG', labelpad=20)

    # Show the plot
    plt.show()

    # Optionally, save the plot
    fig.savefig(f"{name}_heatmap.png")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plot_filtered_dfs = [filtered_df1, filtered_df2, filtered_df3, filtered_df4]
names_filtered_dfs = ["Palenquero_Non_Sequential", "Palenquero_Words_7", "Palenquero_10", "Palenquero_Jundo"]

for i, (f, name) in enumerate(zip(plot_filtered_dfs, names_filtered_dfs), start=1):
    # Create a new figure and axis object
    fig, ax = plt.subplots(figsize=(8, 8))

    # Set the title for the current plot
    ax.set_title(f"Filtered DataFrame: {name}", pad=15)

    # Create a copy of the DataFrame to avoid SettingWithCopyWarning
    f = f.copy()

    # Replace values in 'root_head_match' column
    f['root_head_match'] = f['root_head_match'].replace({0: 'no-match', 1: 'match'})

    # Create the 2D histogram DataFrame
    df_2dhist = pd.DataFrame({
        x_label: grp['UPOSTAG'].value_counts()
        for x_label, grp in f.groupby('root_head_match')
    }).fillna(0)  # Fill NaN with 0 for empty cells

    # Plot the heatmap
    sns.heatmap(df_2dhist, cmap='viridis', annot=True, fmt='g', ax=ax, linewidths=1, linecolor='white', annot_kws={"size": 20, "weight": "bold"})  # Add lines separating sections

    # Set the labels
    ax.set_xlabel('Root-Head Match', labelpad=15)
    ax.set_ylabel('UPOSTAG', labelpad=20)

    # Show the plot
    plt.show()

    # Optionally, save the plot
    fig.savefig(f"{name}_heatmap.png")
```
