```python
# Define the file path
file_path = txt_file

# Read the content of the file
with open(file_path, 'r') as file:
    content = file.read()

# Replace all newline characters with a single space
content = content.replace('\n', '-')

# Write the modified content back to the same file
with open(file_path, 'w') as file:
    file.write(content)

print("Newline characters replaced with a single space successfully.")
```

```python
import re

def split_sentences(text):
    # Split the text into sentences using regular expressions
    sentences = re.split(r'(?<=[.!?])\s+', text)
    return sentences

def main(input_file, output_file):
    with open(input_file, 'r') as f:
        text = f.read()

    sentences = split_sentences(text)

    with open(output_file, 'w') as f:
        for sentence in sentences:
            f.write(sentence + '\n')

if __name__ == "__main__":
    input_file = txt_file  # Specify the input file
    output_file = "output.txt"  # Specify the output file
    main(input_file, output_file)
```
