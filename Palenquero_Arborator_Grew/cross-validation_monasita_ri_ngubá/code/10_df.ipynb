```python
input_csv = '/content/drive/MyDrive/TFM/monasita ri nguba패/statistics_cross_val/cross_validation_original.csv'
output_csv = '/content/drive/MyDrive/TFM/monasita ri nguba패/statistics_cross_val/cross_val_all.csv'
all_cross_val= '/content/drive/MyDrive/TFM/monasita ri nguba패/conllu-validation/all_conllu.csv'
combined_csv_file = '/content/drive/MyDrive/TFM/monasita ri nguba패/statistics_cross_val/combined_cross_val_all.csv'
```

```python
import pandas as pd

# Read the DataFrame from the CSV file
df = pd.read_csv(input_csv)

# Replicate and concatenate the DataFrame 9 times
df_replicated = pd.concat([df] * 10, ignore_index=True)

# Save the resulting DataFrame back to a CSV file
df_replicated.to_csv(output_csv, index=False)
```

```python
import pandas as pd

# Read the first CSV file
csv1 = pd.read_csv(output_csv)

# Read the second CSV file
csv2 = pd.read_csv(all_cross_val)

# Concatenate the two DataFrames side by side (axis=1)
combined_csv = pd.concat([csv1, csv2], axis=1)

# Save the combined DataFrame to a new CSV file
combined_csv.to_csv(combined_csv_file, index=False)
```
