```python
import os
import csv

def extract_root_lines_to_csv(input_file_path, output_file_path):
    headers = ['filename', 'sent_number', 'FORM', 'LEMMA', 'UPOSTAG', 'XPOSTAG', 'FEATS', 'HEAD', 'DEPREL', 'DEPS', 'MISC']
    file_name = os.path.basename(input_file_path)

    with open(input_file_path, 'r', encoding='utf-8') as infile, open(output_file_path, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)

        # Write the headers to the CSV file
        writer.writerow(headers)

        for line in infile:
            # Skip empty lines and comments
            if line.strip() and not line.startswith('#'):
                columns = line.split('\t')
                if len(columns) > 7 and columns[6] == '0':
                    writer.writerow([file_name] + columns[:10])  # Write filename and the first 10 columns

def convert_conllu_folder_to_csv(conllu_folder, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterate over each file in the input folder
    for filename in os.listdir(conllu_folder):
        if filename.endswith('.conllu'):
            input_file_path = os.path.join(conllu_folder, filename)
            output_file_path = os.path.join(output_folder, filename.replace('.conllu', '.csv'))
            extract_root_lines_to_csv(input_file_path, output_file_path)

# Paths to your input CoNLL-U folder and output CSV folder
conllu_folder = conllu_folder
output_folder = csv_folder

# Call the function to convert the folder
convert_conllu_folder_to_csv(conllu_folder, output_folder)
```

```python
import os
import csv

def extract_root_lines_to_csv(input_file_path, writer, file_name):
    for line in open(input_file_path, 'r', encoding='utf-8'):
        # Skip empty lines and comments
        if line.strip() and not line.startswith('#'):
            columns = line.split('\t')
            if len(columns) > 7 and columns[6] == '0':
                writer.writerow([file_name] + columns[:10])  # Write filename and the first 10 columns

def convert_conllu_folder_to_csv(conllu_folder, output_file_path):
    headers = ['filename', 'sent_number', 'FORM', 'LEMMA', 'UPOSTAG', 'XPOSTAG', 'FEATS', 'HEAD', 'DEPREL', 'DEPS', 'MISC']

    # Open the single output CSV file
    with open(output_file_path, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        # Write the headers to the CSV file
        writer.writerow(headers)

        # Iterate over each file in the input folder
        for filename in os.listdir(conllu_folder):
            if filename.endswith('.conllu'):
                input_file_path = os.path.join(conllu_folder, filename)
                extract_root_lines_to_csv(input_file_path, writer, filename)

# Paths to your input CoNLL-U folder and single output CSV file
conllu_folder = conllu_folder
output_file_path = csv_output_file_path

# Call the function to convert the folder and save to one CSV file
convert_conllu_folder_to_csv(conllu_folder, output_file_path)
```
