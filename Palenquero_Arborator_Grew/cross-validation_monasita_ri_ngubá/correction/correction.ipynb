```python
# Assuming your conllu file is named 'input.conllu'
input_file = '/content/drive/MyDrive/TFM/monasita ri ngubaÃÅ/correction/correction.conllu'
output_file = 'filtered_output.conllu'

with open(input_file, 'r', encoding='utf-8') as f_in:
    lines = f_in.readlines()

# Filter lines that start with '# text ='
filtered_lines = [line for line in lines if line.startswith('# text =')]

# Print or write filtered lines to a new file
with open(output_file, 'w', encoding='utf-8') as f_out:
    f_out.writelines(filtered_lines)

print(f"Filtered lines have been written to '{output_file}'")
```
