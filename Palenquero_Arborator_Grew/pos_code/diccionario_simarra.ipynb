```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''

        for page_number in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_number]
            text += page.extract_text()

    return text

# Replace 'your_pdf_file.pdf' with the actual path to your PDF file
pdf_path = '/content/drive/MyDrive/Neología/proyecto/diccionario-de-la-lengua-palenquera_compress.pdf'

try:
    extracted_text = extract_text_from_pdf(pdf_path)
    print(extracted_text[:100])
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import re

def extract_sentences(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()

    pattern = r'\n.*?\).'
    #pattern = r'(\n.*?\)|\.[A-Z].*?\))'
    matches = re.findall(pattern, content)

    return matches

# Replace 'your_file.txt' with the actual filename
filename = '/content/drive/MyDrive/Neología/proyecto/dict_simarra.txt'
selected_sentences = extract_sentences(filename)

for sentence in selected_sentences:
    print(sentence.strip())
```

```python
def save_list_to_txt(lst, file_path):
    with open(file_path, 'w') as file:
        for item in lst:
            file.write(str(item) + '\n')

# File path to save the list
file_path = '/content/drive/MyDrive/Neología/proyecto/dict_simarra.txt'

# Save the list to a text file
save_list_to_txt(selected_sentences, file_path)

print("List saved to", file_path)
```

```python
import pandas as pd
from unidecode import unidecode

# Assuming the delimiter is a comma (','), change it if needed
df_diccionario = pd.read_csv('/content/drive/MyDrive/Neología/proyecto/dict_simarra.txt', delimiter=',,', engine='python')

# Split the 'Skills' column by comma and expand into separate columns
df_diccionario[['Word','POS', 'X']] = df_diccionario['Word,POS'].str.split(',', expand=True)

# Drop the original 'Skills' column if needed
df_diccionario.drop(columns=['Word,POS', 'X'], inplace=True)

# Duplicate column 'A' to column 'B'
df_diccionario['Word2'] = df_diccionario['Word']

# Function to remove accents and make lowercase
def remove_accents_and_lower(word):
    return unidecode(word).lower()

# Apply the function to the 'Words' column
df_diccionario['Word2'] = df_diccionario['Word2'].apply(remove_accents_and_lower)

df_diccionario
```

```python
# Export the DataFrame to a CSV file with a specific file path
file_path = '/content/drive/MyDrive/Neología/proyecto/diccionario_simarra_cleaned.csv'
df_diccionario.to_csv(file_path, index=False)
```

```python
# Function to count cells starting with each letter in a specific column
def count_cells_starting_with_letter_in_column(df, column_name):
    letter_counts = {}
    for cell in df[column_name]:
        if isinstance(cell, str) and len(cell) > 0:
            first_letter = cell[0].upper()
            if first_letter not in letter_counts:
                letter_counts[first_letter] = 1
            else:
                letter_counts[first_letter] += 1
    return letter_counts

# Count cells starting with each letter in a specific column
column_name = 'Word2'  # Change this to the column name you want to analyze
letter_counts = count_cells_starting_with_letter_in_column(df_diccionario, column_name)

# Sort the dictionary by counts in descending order
sorted_counts = sorted(letter_counts.items(), key=lambda x: x[1], reverse=True)

# Print counts of cells starting with each letter in descending order
print(f"Counts of cells starting with each letter in column '{column_name}', from largest to smallest:")
for letter, count in sorted_counts:
    print(f"{letter}: {count}")
```

```python
# Define a function to change the content of the column based on if statements
def change_pos(pos):
    if pos in ["(Adv)", "(Adv.)"]:
        return 'ADV'
    elif pos in ["(Sust)", "( Sust)", "(Sust.)", "(Sust)M"]:
        return 'NOUN'
    elif pos in ["(Verb)", "(Ger)", "(Ger.)", "(Prét)", "(Verboi)", "(Verb imper)", "(Infl. verb)"]:
        return 'VERB'
    #elif pos in ["(Adj.)", "(Adj)"]:
    elif pos in ["(Adj.)", "(Adj)", "(Sust. Adj)", "(Sust. Adj.)"]:
        return 'ADJ'
    elif pos in ["(Prep.)"]:
        return 'ADP'
    elif pos in ["(Interj)", "(Interj.)"]:
        return 'INTJ'
    elif pos in ["(Sust. prop)"]:
        return 'PROPN'
    elif pos in ["(Pron–interr)"]:
        return 'ADV'
    elif pos in ["(Adj. Posesivo)", "(det. plural)"]:
        return 'DET'
    elif pos in ["(Pron. pers)", "(Pron. Pers)", "(Prom. Pers)"]:
        return 'PRON'
    else:
        return pos

# Apply the function to the 'POS' column
pos_simarra['upostag'] = pos_simarra['upostag'].apply(change_pos)
```
