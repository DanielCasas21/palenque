```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''

        for page_number in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_number]
            text += page.extract_text()

    return text

# Replace 'your_pdf_file.pdf' with the actual path to your PDF file
pdf_path = '/content/drive/MyDrive/Neología/proyecto/ocr_solmary.pdf'

try:
    extracted_text = extract_text_from_pdf(pdf_path)
    print(extracted_text[:100])
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import re

# Replace every occurrence of "-\n" with an empty string
cleaned_text = extracted_text.replace("- \n", "")

# Print the result
print(cleaned_text)
```

```python
import re

def extract_sentences(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()

    pattern = r'\n\b.*:.*\n'
    #pattern = r'(\n.*?\)|\.[A-Z].*?\))'
    matches = re.findall(pattern, content)

    return matches

# Replace 'your_file.txt' with the actual filename
filename = '/content/drive/MyDrive/Neología/proyecto/dict_solmary.txt'
selected_sentences = extract_sentences(filename)

for sentence in selected_sentences:
    print(sentence.strip())
```

```python
len(selected_sentences)
```

```python
def save_list_to_txt(lst, file_path):
    with open(file_path, 'w') as file:
        for item in lst:
            file.write(str(item) + '\n')

# File path to save the list
file_path = '/content/drive/MyDrive/Neología/proyecto/dict_solmary.txt'

# Save the list to a text file
save_list_to_txt(selected_sentences, file_path)

print("List saved to", file_path)
```

```python
# Assuming the delimiter is a comma (','), change it if needed
conllu_solmery = pd.read_csv('/content/drive/MyDrive/Neología/proyecto/POS/dict_solmary_conllu.csv', delimiter=',', engine='python')

conllu_solmery_pos = conllu_solmery.loc[~conllu_solmery["UPosTag"].isin(["exp comp"])]

conllu_solmery_pos = conllu_solmery_pos.drop(['Original_POS'], axis=1)

# Changing value in cell (row 1, column 'B') from 6 to 20
conllu_solmery_pos.loc[41, 'feature'] = "Tense=Imp"

conllu_solmery_pos
```

```python
with open('/content/drive/MyDrive/Neología/proyecto/POS/solmery_conllu.txt', 'w') as f:
    for index, row in conllu_solmery.iterrows():
        sent_id = index + 1
        row_string = '	'.join([str(val) for val in row])
        f.write("# sent_id = " + str(sent_id) + "\n")
        f.write("# text = " + str(row['Form']) + ".\n")
        f.write("# title = Diccionario de la Lengua Afro Palenquera - Español\n")
        f.write("# author = Solmery Cásseres Estrada\n")
        f.write("# year = 2005\n")
        f.write("1\t" + row_string + "\n")
        f.write("2\t.\t.\tPUNCT\t_\t_\t1\tpunct\t_\t_\n\n")
```
